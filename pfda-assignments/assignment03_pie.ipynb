```python
# Modules needed for the assignment:
import pandas as pd

import matplotlib.pyplot as plt

from collections import Counter as ct
```

```python
# reading in the data
data = pd.read_csv("https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download")
```

```python
# Source - Andrew Beatty Programming for Data Analytics, Lecture 3.2 Tutorial pandas reading in csv
# Only show the end of the email string
domain = data['Email'].str.split('@').str[1]
domain
```

```python
# Source - https://stackoverflow.com/questions/55852570/how-can-i-get-all-the-unique-categories-within-my-dataframe-using-python

# I wanted to see exactly how many unique domains there were
uc = set(domain)
uc
```

```python
# Source - https://stackoverflow.com/questions/20665232/how-to-plot-the-number-of-times-each-element-is-in-a-list
# This counts the unique categories in domain
count = ct(domain)
count
```

```python
# This separates the int values from the dict, which can be used to plot the data on a piechart

count.values()
```

```python
# Source - https://www.geeksforgeeks.org/data-science/plot-a-pie-chart-in-python-using-matplotlib/
# Source -  https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html
fig = plt.figure (figsize = (10, 10 ))

# The below source was incredibly useful to help me to figure out plot label sizing
# Source - https://stackoverflow.com/questions/7082345/how-to-set-the-labels-size-on-a-pie-chart

# The values are plotted, then labelled outside and the percentage of each wedge is listed
plt.pie(count.values(), labels = count.values(),textprops={'fontsize': 14, 'color' : '#5a270f', 'weight':'bold'}, autopct='%1.1f%%', colors =['#ffbe0b', '#fb5607','#20a39e'])

# The legend is set
# Source - https://matplotlib.org/stable/api/legend_api.html
legend = plt.legend(count.keys(), labelcolor = "#5a270f", shadow = True, loc = "upper left", title = 'Legend', facecolor='#f4f1de', fontsize = 'large', title_fontsize = 'x-large')
# Source - https://matplotlib.org/stable/gallery/text_labels_and_annotations/legend_demo.html
legend.get_title().set_color("#281006")

plt.show()
```
